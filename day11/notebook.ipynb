```python
import string
from pathlib import Path
from collections import defaultdict
import json
import numpy as np
import math
file = open('./input.txt', 'r')

cycle_values = [1]
interesting_cycles = [20,60,100,140,180,220]
current_value = 1

monkeys = []
monkeys_items = {}
monkeys_divisibility = {}
monkeys_operation = {}
monkeys_throws_true = {}
monkeys_throws_false = {}
current_monkey = None

def create_function(op, num):
    if op=='+' and num!='old':
        num = int(num)
        operator = lambda x: x+num
    elif op=='+' and num=='old':
        operator = lambda x: x+x
    elif op=='*' and num!='old':
        num = int(num)
        operator = lambda x: x*num
    else:
        assert op=='*' and num=='old'
        operator = lambda x: x*x
    return operator

for line in file.readlines():
    line = line.replace('\n','')
    if 'Monkey' in line:
        monkey_num = int(line.replace('Monkey ','').replace(':',''))
        current_monkey = monkey_num
        monkeys.append(monkey_num)
    elif 'Starting items' in line:
        nums = line.split(':')[1].split(',')
        nums = [int(num.strip()) for num in nums]
        monkeys_items[current_monkey] = nums
    elif 'Operation:' in line:
        assert 'Operation: new = old ' in line
        line = line.replace('Operation: new = old ','')
        op, num = line.strip().split(' ')
        operator = create_function(op, num)
        monkeys_operation[current_monkey] = operator
    elif 'Test:' in line:
        num = int(line.replace('Test: divisible by ',''))
        monkeys_divisibility[current_monkey] = num
    elif 'If' in line:
        num = int(line.split('monkey')[1].strip())
        if 'true' in line:
            monkeys_throws_true[current_monkey] = num
        else:
            assert 'false' in line
            monkeys_throws_false[current_monkey] = num
```

```python
for monkey in monkeys:
    print(f'considering monkey {monkey}')
    print(f'it has items: {monkeys_items[monkey]}')
    print(f'it has operation that on 2 gives {monkeys_operation[monkey](2)}')
    print(f'it has operation that on 3 gives {monkeys_operation[monkey](3)}')
    print(f'it has divisibility: {monkeys_divisibility[monkey]}')
    print(f'it has throws to: {monkeys_throws_true[monkey]}, {monkeys_throws_false[monkey]}')
    
```
