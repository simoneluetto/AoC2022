```python
import string
from pathlib import Path
from collections import defaultdict
import json
import numpy as np
import math
file = open('./input.txt', 'r')

rock_positions = []
all_x = []
all_y = []
starting_sand = (500,0)

for line in file.readlines():
    line = line.replace('\n','')
    steps = line.split('->')

    steps = [[int(step.strip().split(',')[0]),
            int(step.strip().split(',')[1])]
                for step in steps]
    rock_positions.append(steps)
    all_x.extend([step[0] for step in steps])
    all_y.extend([step[1] for step in steps])

print(f'num rocks = {len(rock_positions)}')

min_x = min(all_x)
max_x = max(all_x)
min_y = min(all_y)
max_y = max(all_y)
print(f'grid x is {min_x}-{max_x}')
print(f'grid y is {min_y}-{max_y}')

cave = np.zeros((max_x+1,max_y+2))
for rock_path in rock_positions:
    assert len(rock_path)>=2
    for i in range(1,len(rock_path)):
        prev = rock_path[i-1]
        next = rock_path[i]
        if prev[0]==next[0]:
            y1 = min(prev[1],next[1])
            y2 = max(prev[1],next[1])
            cave[prev[0], y1:y2+1] = 1
        else:
            assert prev[1]==next[1]
            x1 = min(prev[0],next[0])
            x2 = max(prev[0],next[0])
            cave[x1:x2+1, prev[1]] = 1

for j in range(cave.shape[1]):
    line = f'{j}'
    for i in range(cave.shape[0]):
        if cave[i,j]==1:
            line+='#'
        elif cave[i,j]==2:
            line+='.'
        else:
            line+=' '
    print(line)
```

```python
print(f'starting sand position : {starting_sand}')
print(f'total empty spaces : {np.sum(cave==0)}')
finished = False
num_grains = 0
sand_pos = starting_sand
while not finished:

    if cave[sand_pos[0],sand_pos[1]+1]==0:
        sand_pos = (sand_pos[0],sand_pos[1]+1)
    elif cave[sand_pos[0]-1,sand_pos[1]+1]==0:
        sand_pos = (sand_pos[0]-1,sand_pos[1]+1)
    elif cave[sand_pos[0]+1,sand_pos[1]+1]==0:
        sand_pos = (sand_pos[0]+1,sand_pos[1]+1)
    else:
        assert cave[sand_pos[0],sand_pos[1]]==0
        cave[sand_pos[0],sand_pos[1]] = 2
        num_grains += 1
        sand_pos = starting_sand

    if sand_pos[1]==cave.shape[1]-1:
        finished=True

print(f'total sand is {num_grains}')
print(f'tot sand is {np.sum(cave==2)}')
```
