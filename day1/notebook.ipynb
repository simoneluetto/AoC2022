```python
file = open('./input.txt', 'r')

elf_sums = []
tmp=0
for line in file.readlines():
    line = line.strip('\n')
    if len(line)>0:
        cal = int(line)
        tmp += cal
    else:
        elf_sums.append(tmp)
        tmp=0
elf_sums.sort()
elf_sums = elf_sums[::-1]
print(elf_sums)
print(sum(elf_sums[:3]))
```
